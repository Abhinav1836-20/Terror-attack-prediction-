```python
import gradio as gr
```

```python
 def predict(region, attack_type, weapon_type, target_type, casualties):
    score = 0

    region_risk = {
        "South Asia": 1.0, "Middle East": 0.9, "Africa": 0.8,
        "Americas": 0.6, "Europe": 0.5
    }
    score += region_risk[region]

    attack_risk = {
        "Bombing": 1.2, "Shooting": 1.0, "Assassination": 0.9,
        "Arson": 0.5, "Hostage Situation": 1.1
    }
    score += attack_risk[attack_type]

    weapon_risk = {
        "Explosives": 1.2, "Firearms": 1.0, "Melee": 0.4,
        "Vehicle": 0.6, "Fire": 0.5
    }
    score += weapon_risk[weapon_type]

    target_risk = {
        "Civilians": 1.2, "Military": 0.9, "Government": 1.0,
        "Religious Group": 1.1, "Infrastructure": 0.5
    }
    score += target_risk[target_type]

    if casualties < 5:
        score += 0.2
    elif casualties < 20:
        score += 0.5
    else:
        score += 0.9

    if score < 3:
        severity = "LOW Threat"
    elif score < 4.5:
        severity = "MEDIUM Threat"
    else:
        severity = "HIGH Threat"

    return f"Predicted Terror Attack Severity: {severity}"
```

```python
regions = ["South Asia", "Middle East", "Africa", "Americas", "Europe"]
attacks = ["Bombing", "Shooting", "Assassination", "Arson", "Hostage Situation"]
weapons = ["Explosives", "Firearms", "Melee", "Vehicle", "Fire"]
targets = ["Civilians", "Military", "Government", "Religious Group", "Infrastructure"]
```

```python
iface = gr.Interface(
    fn=predict,
    inputs=[
        gr.Dropdown(regions, label="Region"),
        gr.Dropdown(attacks, label="Attack Type"),
        gr.Dropdown(weapons, label="Weapon Type"),
        gr.Dropdown(targets, label="Target Type"),
        gr.Number(label="Casualties")
    ],
    outputs=gr.Textbox(label="Prediction"),
    title="Balanced Terror Attack Severity Predictor"
)

iface.launch()








```
